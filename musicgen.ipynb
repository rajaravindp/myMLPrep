```python
from transformers import pipeline
import scipy
```

```python
synthesiser = pipeline("text-to-audio", "facebook/musicgen-medium")
```

```python
music = synthesiser("bagpipers at a local pub", forward_params={"do_sample": True, "max_new_tokens": 750}, )
```

```python
scipy.io.wavfile.write("musicgen_out.wav", rate=music["sampling_rate"], data=music["audio"])
```

```python
# Import the wav file and play it
import IPython.display as ipd
ipd.Audio("/content/musicgen_out.wav")
```
