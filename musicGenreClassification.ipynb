# Init FTX for DistilHuBERT

```python
from transformers import AutoFeatureExtractor

model_name = "facebook/hubert-base-ls960"
ftx = AutoFeatureExtractor.from_pretrained(model_name, do_normalize=True, return_attention_mask=True,)
```

```python
ftx
```

# Perform resampling to 16 KHz

```python
# Sampling rate of feature extractor
ftx_sr = ftx.sampling_rate; ftx_sr
```

```python
from datasets import Audio

ds = ds.cast_column("audio", Audio(sampling_rate=ftx_sr))
```

```python
ds['train'][0] # Sampling rate matches ftx_sr
```

# Process resampled audio

```python
# Validate - model accepts new data format
sample = ds['train'][0]['audio']
inputs = ftx(sample['array'], sampling_rate=sample['sampling_rate'])

inputs['input_values'], inputs['attention_mask']
```

```python
# Apply preprocessing to entire data

max_len = 30.0

def preprocess(sample):
  audio_arr = [i['array'] for i in sample['audio']]
  inputs = ftx(
      audio_arr,
      sampling_rate= ftx_sr,
      max_length = int(ftx_sr * max_len),
      truncation = True,
      return_attention_mask = True,
  )
  return inputs

ds_processed = ds.map(
    preprocess,
    batched=True,
    batch_size = 100,
    remove_columns=['audio', 'song_id', 'genre', 'genre_id'],
    num_proc = 2
)
ds_processed
```

```python
# Rename genre field to lab
ds_processed = ds_processed.rename_column('genre_type', 'label')
```

# Fine Tuning

```python
id2label = {v: k for k, v in genre_to_int.items()}
```

```python
from transformers import AutoModelForAudioClassification

num_labs = len(genre_types)
model = AutoModelForAudioClassification.from_pretrained(model_name, num_labels=num_labs, id2label = id2label )
```

```python
# Define compute metrics
import evaluate
import numpy as np

metric = evaluate.load("accuracy")
def compute_metrics(eval_pred):
    predictions = np.argmax(eval_pred.predictions, axis=1)
    return metric.compute(predictions=predictions, references=eval_pred.label_ids)
```

```python
from transformers import TrainingArguments

model_name = model_name.split("/")[-1]
batch_size = 8
gradient_accumulation_steps = 1
num_train_epochs = 10

training_args = TrainingArguments(
    f"{model_name}-finetuned-gtzan",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=5e-5,
    per_device_train_batch_size=batch_size,
    gradient_accumulation_steps=gradient_accumulation_steps,
    per_device_eval_batch_size=batch_size,
    num_train_epochs=num_train_epochs,
    warmup_ratio=0.1,
    logging_steps=5,
    load_best_model_at_end=True,
    metric_for_best_model="accuracy",
    fp16=True,
)
```

```python
model
```

```python
# Init Trainer
from transformers import Trainer

trainer = Trainer(
    model,
    training_args,
    train_dataset=ds_processed["train"],
    eval_dataset=ds_processed["test"],
    tokenizer=ftx,
    compute_metrics=compute_metrics,
)

trainer.train()
```
